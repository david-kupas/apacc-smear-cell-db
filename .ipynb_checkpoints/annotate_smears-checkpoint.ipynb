```python
# Create function to read images into an array

from matplotlib import pyplot as plt
%matplotlib inline
import os

def save_ims_to_array(num_ims, dir_path):
    im_arr = []
    im_idx = 0
    all_im_names = os.listdir(dir_path) 
    im_cnt = min(len(all_im_names), num_ims)
    for im_name in all_im_names:
        im_idx += 1
        
        if im_idx % 10 == 0:
            print("Saving images into array: {:6d}/{:6d}".format(im_idx, im_cnt))
            
        if 'checkpoints' not in im_name:
            im = plt.imread(dir_path + im_name)
            im_arr.append(im)
            
        if im_idx == im_cnt or im_idx == num_ims:
            print("Saving images into array: {:6d}/{:6d}".format(im_idx, im_cnt))
            print("Saving images into array finished.")
            return im_arr
        
```

```python
# Show some example images in a grid

from mpl_toolkits.axes_grid1 import ImageGrid

im_arr = save_ims_to_array(16, output_images_dir_path)

print("Showing images ..")
fig = plt.figure(figsize=(20., 20.))
grid = ImageGrid(fig, 111, 
                 nrows_ncols=(4, 4), # change this to a grid that you desire
                 axes_pad=0.1,
                 )

for ax, im in zip(grid, im_arr):
    ax.imshow(im)

plt.show()
```
